```python
# Loading necessary library
import pickle
import numpy as np
import cv2 as cv
```

```python
# Loading model
pickle_in = open("German_Traffic.p", "rb")
model = pickle.load(pickle_in)
```

```python
# declare size of image and threshold
width = 640
height = 480
threshold = .60
```

```python
# Preprocessing image
def preProcessing(img):
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img = cv.equalizeHist(img)
    img = img /255
    return img
```

```python
# Capturing image using webcam
cap = cv.VideoCapture(0)
cap.set(3, width)
cap.set(4, height)
```

```python
while True:
    success, img_original = cap.read()
    img = np.asarray(img_original)
    img = cv.resize(img, (30, 30))
    #img = preProcessing(img)
    cv.imshow("Webcam Imge", img)
    img = img.reshape(1, 30,30, 3)

    # predict
    class_index = int(model.predict_classes(img))
    # print(class_index)
    predictions = model.predict(img)
    # print(predictions)
    prob_value = np.amax(predictions)
    print(class_index, prob_value)

    if prob_value > threshold:
        cv.putText(img_original, "class: "+str(class_index )+ "  "+ "probability: "+str(prob_value),
                   (50 ,50), cv.FONT_HERSHEY_COMPLEX,
                   1, (0 ,0 ,255), 1)

    cv.imshow('Original Image', img_original)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break
    
```
